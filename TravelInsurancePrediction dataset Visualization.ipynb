```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
frame=pd.read_csv("TravelInsurancePrediction.csv")
frame
```

```python
unnamed_cols = frame.columns.str.contains('Unnamed')

frame_new = frame.drop(frame[frame.columns[unnamed_cols]], axis=1)

frame_new
```

```python
plt.pie(frame_new ["TravelInsurance"].value_counts(), labels=["0", "1"], autopct="%.2f%%")
```

#### From the above pie chart we can notice  that about 64.27% of the people in the dataset do not have travel insurance, while about 35.73% do.

 #### From the above histogram we can notice that the largest numbers of people in this dataset their age between 28 and 29 and there are approximately 500 person. Also,  there a large number of people whose age above 34 years old.

```python
sns.catplot(x='ChronicDiseases' , y='Age',hue='FrequentFlyer',kind='box', data=frame)

```

###### We noticed from the above chart that the group age that didn't travel much, and without chronic diseases is the same group age that didn't travel much, and had chronic diseases ranging between 28 and 32 years, while we noticed a difference in the age group that travels frequently, where the group age that travels frequently, and without chronic diseases is between 27 and 33 years old, and the group age that travels a lot and had chronic diseases is between 26 and 33 years.

###### We concluded that the number of people who travel frequently and had chronic diseases is greater than those without chronic diseases.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=frame, x='Age', hue='EverTravelledAbroad', multiple='stack', kde=True, palette='Reds')
plt.xlabel('Age')
plt.title('Distribution of Age by Travel History')

plt.show()
```

#### The majority of people in the dataset are between the ages of 20 and 40. There are more people who have not traveled abroad than people who have traveled abroad. The number of people who have traveled abroad increases as age decreases.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=frame, x= 'ChronicDiseases', hue='EverTravelledAbroad', multiple='stack', kde=True, palette='Reds')
plt.xlabel('ChronicDiseases')
plt.title('EverTravelledAbroad vs ChronicDiseases')

plt.show()
```

###### We noticed from the above chart that there are more people without chronic diseases than those with chronic diseases. Moreover, the number of people without chronic diseases who have travelled abroad or haven't travelled is more than the other category.
