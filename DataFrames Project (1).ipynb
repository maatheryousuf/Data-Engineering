```python
import pandas as pd
from pandas import Series,DataFrame
```

#### Then we download a TravelInsurancePrediction dataset from kaggle website.  This is the  reference of the dataset: https://www.kaggle.com/datasets/tejashvi14/travel-insurance-prediction-data?resource=download. After that, to read this dataset we use pd.read_csv() function.


```python
frame=pd.read_csv("TravelInsurancePrediction.csv")
frame
```

#### We use shape keyword in order to know the number of rows and columns of the dataset.

```python
frame.shape
```

#### We use info() function to know more information about the dataset like the columns used, and its data type.

```python
frame.info()
```

### To study the dataset, we retrieve the specified columns.

```python
DataFrame(frame,columns=['ID','Age','AnnualIncome','FamilyMembers','EverTravelledAbroad'])
```

#### To merge data, we take two slices(frame[0:4], frame[15:20])  from the dataset.

```python
s =frame[0:4]
s
```

```python
a = frame[15:20]
a
```

#### After that, we use merge() function to merge the data in the column of FamilyMembers in both s and a. 

```python
s.merge(a, on=['FamilyMembers'], how= "inner")

```

### Since the dataset contains a huge number of rows, we used head() function to retrieve the first 10 columns to make it easy for us to analyze the data.

```python
h=frame[0:10]
h
```

#### The groupby() method in Pandas is used to group the rows of a DataFrame by one or more columns. The get_group() method is then used to get the group of rows that match a specific value.

```python
grouped = h.groupby(h.GraduateOrNot)
h_new = grouped.get_group("Yes")
h_new
```

#### To Retrieve rows by index we can use loc key word and specify the index of each row you want.

```python
h.loc[[3,8,5]]
```

#### The code h[h['AnnualIncome']>800000] selects all rows in the h DataFrame where the AnnualIncome column is greater than 800,000. The output of the h[h['AnnualIncome']>800000] expression will be a new DataFrame with only the rows that match the condition.

```python
h[h['AnnualIncome']>800000]
```

#### Adding new column with its values in the dataFrame . So h is the name of dataframe, 'name' is the name of the new column, and the other values will be added to 'name' column.

```python
h['Name'] = ['Maather', 'Rana', 'Yousuf', 'Rashid', 'Saleh', 'Muzna', 'Fatma', 'Khadija', 'Sheikha', 'Noor']
h
```

#### In ordeder to add new row in Dataframe we can use  the .loc[] indexing method. The .loc[] method is used to access Dataframe elements by label, and it supports adding new rows as well as creating copies of existing ones. The following code represents how to add a Dataframe row at the end.

```python
h.loc[len(h.index)] = [10, 20, 'MS', 'yes', 500000, 2, 0, 'No', 'No',0,'Ahed']
h
```

### In order to delete any column from the dataset we can use del keyword and specify the dataframe name column name.

```python
del h['Name']
h
```

### In order to delete any row from the dataset we can use drop function and specify the index of the row .

```python
h.drop([0, 5])
```

 #### The sort_values() method in Pandas is used to sort  DataFrame by one or more columns. The by parameter specifies the column or columns that you want to sort by.  The h variable is a Pandas Series. The sort_values() method will sort the DataFrame by the Age column in ascending order. The output of the sort_values() method will be a new DataFrame with the same values as the original DataFrame, but with the elements sorted by the Age column.

```python
h.sort_values(by=['Age'])
```
